```python
import mysql.connector as connector
```

```python
connection = connector.connect(user='your_user_name',password='your_user_password',db='little_lemon_db')
cursor = connection.cursor()
```

```python
show_tables_query = '''SHOW TABLES;'''
cursor.execute(show_tables_query)
print(list(cursor.column_names))
results = cursor.fetchall()
for i in results:
    print(i)
```

```python
join_query = '''SELECT 
CONCAT(customers.FirstName,' ',customers.LastName) as "FullName",
customers.ContactNumber, orders.TotalCost
from customers INNER JOIN orders
ON customers.CustomerID = orders.CustomerID
WHERE TotalCost > 60;'''

cursor.execute(join_query)
print(list(cursor.column_names))
r = cursor.fetchall()
for i in r:
    print((i[0],i[1],float(i[2])))
```
